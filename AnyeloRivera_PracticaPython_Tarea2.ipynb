## Ejercicio 16: DataFrame de estudiantes

**Enunciado:** crear un DataFrame con al menos 10 estudiantes que incluya nombre, carrera, edad, promedio y ciudad. Explorar los datos, calcular el promedio general, filtrar y ordenar.

**Logica:** creo el DataFrame con un diccionario de listas. Uso `.head()` para ver las primeras filas, `.info()` para la estructura general, `.mean()` para el promedio, filtrado con condicion y `.sort_values()` para ordenar.

```python
import pandas as pd

datos_estudiantes = {
    "nombre":  ["Ana Lopez", "Luis Reyes", "Maria Torres", "Carlos Paz",
                "Sofia Mejia", "Jorge Diaz", "Valeria Cruz", "Diego Morales",
                "Fernanda Ruiz", "Andres Vega"],
    "carrera": ["Sistemas", "Contabilidad", "Sistemas", "Derecho",
                "Sistemas", "Administracion", "Derecho", "Contabilidad",
                "Sistemas", "Administracion"],
    "edad":    [20, 22, 19, 21, 23, 20, 22, 21, 19, 24],
    "promedio":[88.5, 75.0, 92.3, 68.0, 85.5, 71.2, 60.8, 80.0, 95.1, 77.4],
    "ciudad":  ["Tegucigalpa", "San Pedro Sula", "Tegucigalpa", "La Ceiba",
                "Comayagua", "Tegucigalpa", "San Pedro Sula", "La Ceiba",
                "Tegucigalpa", "Comayagua"]
}

df_est = pd.DataFrame(datos_estudiantes)

print("Primeras 5 filas:")
print(df_est.head())

print("\nInformacion general del DataFrame:")
print(df_est.info())

print(f"\nPromedio general de calificaciones: {df_est['promedio'].mean():.2f}")

print("\nEstudiantes con promedio mayor o igual a 80:")
print(df_est[df_est["promedio"] >= 80])

print("\nEstudiantes ordenados de mayor a menor promedio:")
print(df_est.sort_values("promedio", ascending=False))
```

## Ejercicio 17: Analisis de ventas con pandas

**Enunciado:** crear un DataFrame de ventas con al menos 15 registros que incluya fecha, vendedor, producto, categoria, cantidad y precio unitario. Calcular TotalVenta, analizar por vendedor, producto y categoria.

**Logica:** la columna TotalVenta la creo multiplicando las columnas cantidad y precio. Uso `.groupby()` con `.sum()` para agrupar por vendedor. Para el producto mas vendido uso `.idxmax()` que devuelve el indice del valor maximo. Con `.sort_values()` ordeno de mayor a menor.

```python
import pandas as pd
datos_ventas = {
    "fecha":      ["2024-01-05","2024-01-07","2024-01-08","2024-01-10",
                   "2024-01-12","2024-01-14","2024-01-15","2024-01-17",
                   "2024-01-18","2024-01-20","2024-01-21","2024-01-22",
                   "2024-01-24","2024-01-25","2024-01-28"],
    "vendedor":   ["Carlos","Ana","Carlos","Maria","Ana","Luis","Maria",
                   "Carlos","Ana","Luis","Carlos","Maria","Ana","Luis","Carlos"],
    "producto":   ["Laptop","Mouse","Teclado","Monitor","Laptop","Auriculares",
                   "Mouse","Monitor","Teclado","Laptop","Auriculares","Laptop",
                   "Monitor","Mouse","Teclado"],
    "categoria":  ["Computacion","Accesorios","Accesorios","Computacion",
                   "Computacion","Accesorios","Accesorios","Computacion",
                   "Accesorios","Computacion","Accesorios","Computacion",
                   "Computacion","Accesorios","Accesorios"],
    "cantidad":   [1, 3, 2, 1, 2, 4, 5, 1, 3, 1, 2, 1, 2, 6, 2],
    "precio":     [15000, 350, 500, 8000, 15000, 800, 350, 8000,
                   500, 15000, 800, 15000, 8000, 350, 500]
}

df_ven = pd.DataFrame(datos_ventas)

# columna TotalVenta
df_ven["TotalVenta"] = df_ven["cantidad"] * df_ven["precio"]
print("DataFrame con columna TotalVenta:")
print(df_ven.head(5))

print("\nTotal vendido por vendedor:")
print(df_ven.groupby("vendedor")["TotalVenta"].sum().reset_index().sort_values("TotalVenta", ascending=False))

print("\nProducto mas vendido (por cantidad):")
mas_vendido = df_ven.groupby("producto")["cantidad"].sum()
print(f"  {mas_vendido.idxmax()} con {mas_vendido.max()} unidades")

print("\nCategoria con mayor ingreso:")
cat_ingreso = df_ven.groupby("categoria")["TotalVenta"].sum()
print(f"  {cat_ingreso.idxmax()} con L.{cat_ingreso.max():,}")

print("\nVentas ordenadas de mayor a menor:")
print(df_ven.sort_values("TotalVenta", ascending=False)[["vendedor","producto","cantidad","precio","TotalVenta"]])
```

## Ejercicio 18: Limpieza de datos con pandas

**Enunciado:** crear un DataFrame con datos incompletos de clientes, identificar y contar valores nulos, reemplazar ciudades vacias, eliminar registros sin correo y mostrar el DataFrame limpio.

**Logica:** uso `None` para simular datos faltantes. `.isnull()` devuelve True/False por cada celda, `.sum()` los cuenta. `.fillna()` reemplaza los nulos de una columna con un valor. `.dropna(subset=["columna"])` elimina filas donde esa columna es nula.

```python
import pandas as pd
import numpy as np

datos_clientes = {
    "nombre":    ["Juan Perez", "Maria Gomez", "Carlos Ruiz", "Ana Lima",
                  "Luis Mora", "Sofia Diaz", "Pedro Vega", "Laura Cruz"],
    "edad":      [28, None, 35, 22, None, 30, 45, 27],
    "correo":    ["juan@mail.com", "maria@mail.com", None, "ana@mail.com",
                  "luis@mail.com", None, "pedro@mail.com", "laura@mail.com"],
    "ciudad":    ["Tegucigalpa", None, "San Pedro Sula", None,
                  "La Ceiba", "Tegucigalpa", None, "Comayagua"],
    "telefono":  ["9999-0001", "9999-0002", None, "9999-0004",
                  None, "9999-0006", "9999-0007", None]
}

df_cli = pd.DataFrame(datos_clientes)
print("DataFrame original:")
print(df_cli)

print("\nValores nulos por columna:")
print(df_cli.isnull().sum())

# reemplazar ciudades vacias
df_cli["ciudad"] = df_cli["ciudad"].fillna("No especificada")
print("\nDespues de rellenar ciudades vacias:")
print(df_cli[["nombre","ciudad"]])

# eliminar registros sin correo
df_cli = df_cli.dropna(subset=["correo"])
print("\nDataFrame limpio (sin registros sin correo):")
print(df_cli.reset_index(drop=True))
```

## Ejercicio 19: Analisis de prestamos

**Enunciado:** crear un DataFrame con solicitudes de prestamo, calcular montos promedio, contar por estado, filtrar aprobados y por monto, y agrupar por estado.

**Logica:** uso `.mean()` para el promedio, `.value_counts()` para contar por estado. Los filtros son condiciones directas sobre columnas. `.groupby()` con `.mean()` calcula el promedio de monto por cada estado de solicitud.

```python
import pandas as pd
datos_prestamos = {
    "cliente":         ["Ana Torres", "Luis Mejia", "Maria Paz", "Carlos Vega",
                        "Sofia Ruiz", "Jorge Lima", "Valeria Cruz", "Diego Reyes",
                        "Fernanda Lopez", "Andres Diaz"],
    "monto_solicitado":[45000, 75000, 30000, 120000, 55000,
                        20000, 90000, 48000, 65000, 15000],
    "plazo_meses":     [24, 36, 12, 48, 24, 12, 36, 24, 36, 12],
    "ingreso_mensual": [18000, 25000, 12000, 40000, 20000,
                        9000, 35000, 17000, 22000, 8000],
    "estado":          ["Aprobado", "Aprobado", "Rechazado", "Aprobado",
                        "En revision", "Rechazado", "Aprobado", "En revision",
                        "Aprobado", "Rechazado"]
}

df_pre = pd.DataFrame(datos_prestamos)
print("DataFrame de prestamos:")
print(df_pre)

print(f"\nMonto promedio solicitado: L.{df_pre['monto_solicitado'].mean():,.2f}")

print("\nConteo de solicitudes por estado:")
print(df_pre["estado"].value_counts())

print("\nPrestamos aprobados:")
print(df_pre[df_pre["estado"] == "Aprobado"])

print("\nClientes con monto mayor a L.50,000:")
print(df_pre[df_pre["monto_solicitado"] > 50000])

print("\nPromedio de monto solicitado por estado:")
print(df_pre.groupby("estado")["monto_solicitado"].mean().reset_index())
```

## Ejercicio 20: Mini analisis de datos tipo empresa

**Enunciado:** crear un DataFrame de empleados con nombre, departamento, cargo, salario, anos de experiencia y ciudad. Analizar estadisticas, agrupar, ordenar y crear columna de nivel de experiencia.

**Logica:** `.describe()` genera estadisticas de salario automaticamente. `.idxmax()` localiza el indice del salario mas alto y con `.loc[]` accedo a toda esa fila. La columna NivelExperiencia la creo con `.apply()` y un `lambda` que evalua los anos de experiencia con condiciones anidadas.

```python
import pandas as pd
datos_empleados = {
    "nombre":      ["Ana Lopez", "Luis Reyes", "Maria Torres", "Carlos Paz",
                    "Sofia Mejia", "Jorge Diaz", "Valeria Cruz", "Diego Morales",
                    "Fernanda Ruiz", "Andres Vega", "Carmen Soto", "Miguel Perez"],
    "departamento":["Tecnologia", "Ventas", "Tecnologia", "RRHH",
                    "Ventas", "Tecnologia", "RRHH", "Ventas",
                    "Tecnologia", "RRHH", "Ventas", "Tecnologia"],
    "cargo":       ["Desarrolladora", "Ejecutivo", "Analista", "Reclutador",
                    "Gerente Ventas", "Desarrollador", "Coordinadora", "Ejecutivo",
                    "DBA", "Director RRHH", "Ejecutiva", "DevOps"],
    "salario":     [28000, 22000, 25000, 18000, 40000, 30000, 21000,
                    20000, 32000, 45000, 23000, 35000],
    "experiencia": [3, 2, 5, 1, 8, 6, 2, 1, 7, 10, 3, 9],
    "ciudad":      ["Tegucigalpa", "San Pedro Sula", "Tegucigalpa", "La Ceiba",
                    "Tegucigalpa", "San Pedro Sula", "Tegucigalpa", "La Ceiba",
                    "Comayagua", "Tegucigalpa", "San Pedro Sula", "Tegucigalpa"]
}

df_emp = pd.DataFrame(datos_empleados)

print("Primeros registros:")
print(df_emp.head())

print("\nEstadisticas generales de salarios:")
print(df_emp["salario"].describe())

print("\nSalario promedio por departamento:")
print(df_emp.groupby("departamento")["salario"].mean().reset_index().sort_values("salario", ascending=False))

print("\nEmpleado con salario mas alto:")
idx = df_emp["salario"].idxmax()
print(df_emp.loc[idx])

print("\nEmpleados con mas de 5 anos de experiencia:")
print(df_emp[df_emp["experiencia"] > 5])

print("\nEmpleados ordenados por salario (mayor a menor):")
print(df_emp.sort_values("salario", ascending=False)[["nombre","cargo","salario"]].reset_index(drop=True))

def nivel(exp):
    if exp <= 2:
        return "Junior"
    elif exp <= 5:
        return "Intermedio"
    else:
        return "Senior"

df_emp["NivelExperiencia"] = df_emp["experiencia"].apply(nivel)
print("\nDataFrame con columna NivelExperiencia:")
print(df_emp[["nombre","experiencia","NivelExperiencia"]])

print("\nConteo por nivel de experiencia:")
print(df_emp["NivelExperiencia"].value_counts())
```
