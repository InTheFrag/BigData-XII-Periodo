## Que datos genera CRIOHO?

CRIOHO es un centro de rehabilitacion fisica que brinda servicios de terapia y masajes terapeuticos a personas con discapacidades. Atiende aproximadamente 100 pacientes por dia y cuenta con registros desde antes del año 2009.

**Datos de cada paciente**
- Nombre completo, edad, genero
- Diagnostico medico o tipo de discapacidad
- Medico referente que lo envio al centro
- Fecha de ingreso al programa

**Datos de cada cita o sesion**
- Fecha y hora de la cita
- Terapeuta asignado
- Tipo de terapia (fisioterapia, masaje, hidroterapia, etc.)
- Estado de la cita: atendida, cancelada o reprogramada

**Datos de evolucion clinica**
- Observaciones del terapeuta por sesion
- Nivel de avance del paciente
- Numero de sesiones completadas vs. las prescritas

**Datos administrativos**
- Pagos realizados o pendientes
- Cobertura de seguro medico o IHSS
- Referencias medicas recibidas

En cuanto al volumen, se estima que desde 2009 el centro ha acumulado mas de 400,000 registros de citas unicamente. En cuanto a la frecuencia, los datos se generan de forma continua durante cada jornada laboral, los cinco dias de la semana.

```python
import pandas as pd

datos_volumen = {
    "Concepto": [
        "Pacientes atendidos por dia",
        "Dias laborales por ano",
        "Anos en operacion (desde 2009)",
        "Total de citas acumuladas (estimado)",
        "Registros totales estimados (x8 campos por cita)"
    ],
    "Valor estimado": ["100", "250", "16 anos", "400,000", "3,200,000"]
}

df_volumen = pd.DataFrame(datos_volumen)
print("Estimacion de volumen de datos - CRIOHO")
df_volumen
```

## Quien los usa?

Los datos generados por CRIOHO son utilizados por distintos actores, tanto dentro como fuera del centro:

Terapeutas y fisioterapeutas son quienes recopilan los datos directamente durante cada sesion. Consultan el historial del paciente para saber en que punto del tratamiento se encuentra.

Medicos referentes externos son los profesionales que recomendaron al paciente asistir a CRIOHO. Necesitan acceder a los datos de evolucion para saber si el tratamiento esta funcionando.

El personal administrativo procesa los datos de agenda, pagos y disponibilidad de terapeutas para que el centro funcione ordenadamente.

Los pacientes y sus familiares pueden consultar su propio historial, proximas citas y el progreso reportado.

Seguros medicos e IHSS acceden a los datos de diagnostico y sesiones realizadas para autorizar la cobertura de los tratamientos.

```python
datos_actores = {
    "Actor": [
        "Terapeutas / Fisioterapeutas",
        "Medicos referentes externos",
        "Administracion del centro",
        "Pacientes y familiares",
        "Seguros medicos / IHSS"
    ],
    "Rol": [
        "Recopilan y consultan datos en cada sesion",
        "Consumen datos para dar seguimiento al tratamiento",
        "Procesan datos de agenda, pagos y recursos",
        "Consultan su propio historial y citas",
        "Verifican sesiones para autorizar cobertura"
    ],
    "Datos que consultan": [
        "Historial clinico, plan de tratamiento, sesiones completadas",
        "Evolucion del paciente, numero de sesiones, avance",
        "Citas del dia, pagos, disponibilidad de terapeutas",
        "Proximas citas, progreso reportado",
        "Diagnostico, sesiones realizadas, costos"
    ]
}

df_actores = pd.DataFrame(datos_actores)
print("Actores que utilizan los datos de CRIOHO")
df_actores
```

## Que problema resuelve?

Organizacion y continuidad del tratamiento. Con mas de 100 pacientes diarios, es imposible llevar el control manual de cada sesion. El analisis de datos permite saber exactamente en que sesion va cada paciente y si esta siguiendo el tratamiento de manera constante.

Seguimiento clinico a distancia. Los medicos referentes no siempre pueden visitar al paciente en el centro. Con acceso a los datos de evolucion, pueden tomar decisiones medicas informadas sin desplazarse.

Optimizacion de la agenda y los recursos. Al analizar las horas con mayor demanda o los terapeutas con mayor carga, el centro puede distribuir mejor sus recursos.

Historial clinico a largo plazo. Con datos desde antes de 2009, es posible identificar patrones: que tipos de discapacidades son mas frecuentes o cuantas sesiones necesita un paciente en promedio para mejorar.

```python
datos_pacientes = {
    "ID Paciente": [f"P{str(i).zfill(3)}" for i in range(1, 11)],
    "Tipo de terapia": [
        "Fisioterapia","Masaje terapeutico","Hidroterapia",
        "Fisioterapia","Terapia ocupacional","Masaje terapeutico",
        "Fisioterapia","Hidroterapia","Fisioterapia","Terapia ocupacional"
    ],
    "Sesiones prescritas": [20,12,15,20,10,12,20,15,20,10],
    "Sesiones completadas": [18,12,9,5,10,7,20,14,11,3],
    "Estado": [
        "En progreso","Completado","En progreso","En progreso","Completado",
        "En progreso","Completado","En progreso","En progreso","En progreso"
    ]
}

df_pacientes = pd.DataFrame(datos_pacientes)
df_pacientes["% Avance"] = (df_pacientes["Sesiones completadas"] / df_pacientes["Sesiones prescritas"] * 100).round(1).astype(str) + "%"

print("Seguimiento de pacientes - ejemplo simulado CRIOHO")
df_pacientes
```

```python
resumen = df_pacientes.groupby("Tipo de terapia").agg(
    Total_pacientes=("ID Paciente", "count"),
    Sesiones_prescritas_total=("Sesiones prescritas", "sum"),
    Sesiones_completadas_total=("Sesiones completadas", "sum")
).reset_index()

resumen["% Cumplimiento"] = (resumen["Sesiones_completadas_total"] / resumen["Sesiones_prescritas_total"] * 100).round(1).astype(str) + "%"

print("Resumen por tipo de terapia")
resumen
```

## Por que no bastaria Excel?

Excel es util para pequenos volumenes de datos, pero en CRIOHO presenta limitaciones importantes explicadas por las **4 Vs del Big Data**:

**Volumen.** Excel tiene un limite practico de rendimiento alrededor de las 100,000 filas. CRIOHO ha acumulado mas de 400,000 registros de citas unicamente. Un archivo de ese tamano tardaria minutos en abrirse.

**Velocidad.** Excel no permite que multiples personas editen el mismo archivo al mismo tiempo sin riesgo de conflictos. En CRIOHO, varios terapeutas y el personal administrativo necesitan registrar y consultar informacion simultaneamente durante la jornada.

**Variedad.** Los datos de CRIOHO son de tipos muy distintos: texto clinico, fechas, diagnosticos, pagos, referencias medicas. Relacionar todas estas tablas en Excel requeriria formulas muy complejas y propensas a errores.

**Veracidad.** Un error en una formula de Excel puede modificar datos clinicos sin dejar rastro. En un centro medico, la integridad del historial del paciente es critica y Excel no ofrece mecanismos de auditoria.

```python
datos_comp = {
    "Criterio": [
        "Capacidad de almacenamiento",
        "Acceso simultaneo de varios usuarios",
        "Velocidad con grandes volumenes",
        "Analisis de datos historicos (desde 2009)",
        "Seguridad e integridad clinica",
        "Generacion de reportes automaticos",
        "Conexion con sistemas externos (IHSS, seguros)"
    ],
    "Excel": [
        "Limitado (~100k filas practicas)",
        "No - genera conflictos al editar",
        "Lento con +50k registros",
        "Muy dificil de gestionar",
        "Sin control de auditoria",
        "Manual, cada vez que se necesita",
        "No tiene integracion nativa"
    ],
    "Big Data / Base de datos": [
        "Millones de registros sin problemas",
        "Si - disenado para multiples usuarios",
        "Rapido con indices y consultas optimizadas",
        "Consultas sobre cualquier rango de fechas",
        "Auditoria completa de cambios",
        "Automatico y programable",
        "Integrable con APIs y sistemas externos"
    ]
}

df_comp = pd.DataFrame(datos_comp)
print("Comparativa: Excel vs Big Data para CRIOHO")
df_comp
```
