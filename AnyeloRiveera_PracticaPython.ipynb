## Ejercicio 1: Clasificacion de edad

**Enunciado:** determinar si una persona es nino, adolescente, adulto o adulto mayor segun su edad.

**Logica:** se evalua la edad usando condiciones `if`, `elif` y `else`. El primer rango que se cumple determina la categoria.

```python
def clasificar_edad(edad):
    if edad < 13:
        return "Nino"
    elif edad <= 17:
        return "Adolescente"
    elif edad <= 59:
        return "Adulto"
    else:
        return "Adulto mayor"

edad = 21
categoria = clasificar_edad(edad)
print(f"Edad: {edad} anos")
print(f"Categoria: {categoria}")
```

## Ejercicio 2: Conversor de temperatura

```python
celsius = 36.5
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius} grados Celsius equivalen a {fahrenheit:.2f} grados Fahrenheit")
```

## Ejercicio 3: Calculadora basica

**Enunciado:** realizar operaciones de suma, resta, multiplicacion y division entre dos numeros, validando que no se divida entre cero.

**Logica:** defino los dos numeros y la operacion como variables. Uso if/elif para cada caso y dentro de la division verifico si el segundo numero es cero antes de operar.

```python
num1 = 10
num2 = 0
operacion = "division"

if operacion == "suma":
    resultado = num1 + num2
    print(f"{num1} + {num2} = {resultado}")
elif operacion == "resta":
    resultado = num1 - num2
    print(f"{num1} - {num2} = {resultado}")
elif operacion == "multiplicacion":
    resultado = num1 * num2
    print(f"{num1} x {num2} = {resultado}")
elif operacion == "division":
    if num2 == 0:
        print("Error: no se puede dividir entre cero")
    else:
        resultado = num1 / num2
        print(f"{num1} / {num2} = {resultado}")
```

**Enunciado:** determinar si un numero entero es par o impar y mostrar el resultado.

## Ejercicio 6: Contador de vocales

**Enunciado:** contar cuantas vocales tiene una frase, sin importar si son mayusculas o minusculas.

## Ejercicio 7: Promedio de calificaciones

**Enunciado:** calcular el promedio de 5 calificaciones e indicar si el estudiante aprobo o reprobo (aprobado con promedio >= 70).

**Logica:** uso `sum()` para sumar la lista y divido entre `len()` para obtener el promedio. Luego comparo con 70 para mostrar el estado.

```python
calificaciones = [85, 72, 90]

total = sum(calificaciones)
promedio = total / len(calificaciones)

print(f"Calificaciones: {calificaciones}")
print(f"Suma total: {total}")
print(f"Promedio: {promedio:.1f}")

if promedio >= 70:
    print("Estado: Aprobado")
else:
    print("Estado: Reprobado")
```

## Ejercicio 8: Lista de numeros positivos y negativos

**Enunciado:** contar cuantos numeros positivos, negativos y ceros hay en una lista de 10 numeros.

**Logica:** recorro la lista con un `for` y con condiciones `if/elif/else` sumo a cada contador segun corresponda. Al final muestro el resumen.

```python
numeros = [4, -2, 0, 7, -5, 3, 0, -1, 9, -8]

positivos = 0
negativos = 0
ceros = 0

for n in numeros:
    if n > 0:
        positivos += 1
    elif n < 0:
        negativos += 1
    else:
        ceros += 1

print(f"Numeros: {numeros}")
print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Ceros: {ceros}")
```

## Ejercicio 9: Mayor y menor de una lista

**Enunciado:** encontrar el numero mayor y el menor de una lista y calcular la diferencia entre ambos.

**Logica:** uso las funciones `max()` y `min()` que Python tiene incorporadas. `max()` recorre la lista internamente y devuelve el valor mas alto, `min()` hace lo mismo con el mas bajo. La diferencia es simplemente mayor menos menor.

```python
numeros = [15, 3, 42, 8, 27, 1, 55, 19]

mayor = max(numeros)
menor = min(numeros)
diferencia = mayor - menor

print(f"Lista: {numeros}")
print(f"Numero mayor: {mayor}")
print(f"Numero menor: {menor}")
print(f"Diferencia entre mayor y menor: {diferencia}")
```

## Ejercicio 11: Funcion para calcular descuento

**Enunciado:** crear una funcion que reciba precio original y porcentaje de descuento, y retorne el precio final.

**Logica:** la funcion recibe dos parametros. Calcula el monto del descuento (precio * porcentaje / 100), lo resta al precio original y retorna ese valor. Lo que entra son dos numeros y lo que sale es un numero con el precio ya descontado.

```python
def calcular_descuento(precio_original, porcentaje):
    descuento = precio_original * porcentaje / 100
    precio_final = precio_original - descuento
    return precio_final

precio = 100
descuento_aplicar = 15

resultado = calcular_descuento(precio, descuento_aplicar)
print(f"Precio original: L. {precio:.2f}")
print(f"Descuento: {descuento_aplicar}%")
print(f"Precio final: L. {resultado:.2f}")
```

## Ejercicio 12: Validacion de contrasena

**Enunciado:** validar si una contrasena tiene al menos 8 caracteres, un numero, una mayuscula y una minuscula.

**Logica:** verifico cada condicion por separado usando funciones como `isupper()`, `islower()` e `isdigit()`. La funcion `any()` revisa si al menos un caracter de la cadena cumple la condicion. Al final combino todo con `and`.

```python
contrasena = "Admin1234"

tiene_longitud = len(contrasena) >= 8
tiene_numero = any(c.isdigit() for c in contrasena)
tiene_mayuscula = any(c.isupper() for c in contrasena)
tiene_minuscula = any(c.islower() for c in contrasena)

print(f"Contrasena: {contrasena}")
print(f"  Longitud minima (8): {tiene_longitud}")
print(f"  Contiene numero: {tiene_numero}")
print(f"  Contiene mayuscula: {tiene_mayuscula}")
print(f"  Contiene minuscula: {tiene_minuscula}")

if tiene_longitud and tiene_numero and tiene_mayuscula and tiene_minuscula:
    print("Resultado: Contrasena VALIDA")
else:
    print("Resultado: Contrasena NO valida")
```

## Ejercicio 13: Sistema simple de inventario

**Enunciado:** crear una lista de productos con codigo, nombre, precio y cantidad. Mostrarlos, buscar por codigo y calcular el valor total.

**Logica:** cada producto es un diccionario dentro de una lista. Para buscar recorro la lista comparando el codigo. El valor total lo calculo multiplicando precio por cantidad de cada producto y sumando todo.

```python
inventario = [
    {"codigo": "A001", "nombre": "Laptop", "precio": 15000, "cantidad": 5},
    {"codigo": "B002", "nombre": "Mouse", "precio": 350, "cantidad": 20},
    {"codigo": "C003", "nombre": "Teclado", "precio": 500, "cantidad": 15},
    {"codigo": "D004", "nombre": "Monitor", "precio": 8000, "cantidad": 8}
]

print(" Inventario completo ")
for p in inventario:
    print(f"  [{p['codigo']}] {p['nombre']} - L.{p['precio']:,} x{p['cantidad']} unidades")

buscar = "B002"
print(f"\nBusqueda: codigo {buscar}")
for p in inventario:
    if p["codigo"] == buscar:
        print(f"  Encontrado: {p['nombre']} a L.{p['precio']:,}")

total = sum(p["precio"] * p["cantidad"] for p in inventario)
print(f"\nValor total del inventario: L.{total:,}")
```

## Ejercicio 14: Registro de ventas

**Enunciado:** crear una lista de ventas con cliente, producto, cantidad y precio. Calcular total por venta, total general e identificar el cliente con la compra mas alta.

**Logica:** el total de cada venta es cantidad x precio. Para el total general sumo todos. Para encontrar la compra mas alta uso `max()` con un `lambda` que le dice por cual campo comparar.

```python
ventas = [
    {"cliente": "Ana Reyes", "producto": "Laptop", "cantidad": 1, "precio": 15000},
    {"cliente": "Carlos Paz", "producto": "Mouse", "cantidad": 3, "precio": 350},
    {"cliente": "Maria Torres", "producto": "Monitor", "cantidad": 2, "precio": 8000},
    {"cliente": "Luis Gomez", "producto": "Teclado", "cantidad": 1, "precio": 500}
]

print(" Detalle de ventas ")
for v in ventas:
    total_venta = v["cantidad"] * v["precio"]
    print(f"  {v['cliente']}: {v['producto']} x{v['cantidad']} = L.{total_venta:,}")

total_general = sum(v["cantidad"] * v["precio"] for v in ventas)
print(f"\nTotal general vendido: L.{total_general:,}")

mejor = max(ventas, key=lambda v: v["cantidad"] * v["precio"])
print(f"Cliente con compra mas alta: {mejor['cliente']} (L.{mejor['cantidad'] * mejor['precio']:,})")
```

## Ejercicio 15: Limpieza basica de texto

**Enunciado:** limpiar una frase eliminando espacios extremos, convirtiendo a minusculas, quitando espacios dobles y contando palabras.

**Logica y metodos:**
- `strip()`: elimina espacios en blanco al inicio y al final de la cadena
- `lower()`: convierte todo el texto a minusculas
- `replace()`: reemplaza una cadena por otra, aqui uso un while para seguir reemplazando hasta que no queden espacios dobles
- `split()`: divide el texto en una lista usando el espacio como separador, y `len()` me dice cuantos elementos tiene esa lista

```python
frase = "  Hola   Mundo  esto  es  una  PRUEBA  "

print(f"Texto original:       '{frase}'")

frase = frase.strip()
print(f"Despues de strip():   '{frase}'")

frase = frase.lower()
print(f"Despues de lower():   '{frase}'")

while "  " in frase:
    frase = frase.replace("  ", " ")
print(f"Sin espacios dobles:  '{frase}'")

palabras = frase.split()
print(f"Numero de palabras:   {len(palabras)}")
```
